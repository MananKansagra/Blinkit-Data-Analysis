```python
import pandas as pd
```

```python
df = pd.read_csv("data/blinkit_inventory.csv")
df
```

```python
# Group by 'product_id' and sum the other columns
grouped_df = df.groupby('product_id', as_index=False).agg({
    'stock_received': 'sum',
    'damaged_stock': 'sum'
})

# Calculate damaged stock percentage
grouped_df['damaged_percentage'] = (grouped_df['damaged_stock'] / grouped_df['stock_received']) * 100

# Optional: Round the percentage to 2 decimal places
grouped_df['damaged_percentage'] = grouped_df['damaged_percentage'].round(2)

# Display result
print(grouped_df)
```
