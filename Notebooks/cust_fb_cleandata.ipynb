```python
import pandas as pd

# ðŸ”¹ Step 1: Load your dataset from file path
# Replace 'your_file_path.csv' with your actual path (CSV or Excel)
df = pd.read_csv('data/blinkit_customer_feedback.csv')    # or pd.read_excel('your_file_path.xlsx')

# ðŸ”¹ Step 2: Define the mapping from rating to experience
rating_map = {
    1: "Very Poor",
    2: "Poor",
    3: "Average",
    4: "Good",
    5: "Excellent"
}

# ðŸ”¹ Step 3: Create the new column 'experience' based on 'rating'
df['experience'] = df['rating'].map(rating_map)
# Save the DataFrame to Excel file
df.to_csv("experience.csv", index=False)
df

```
