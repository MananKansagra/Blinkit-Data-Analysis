```python
import matplotlib.pyplot as plt
x = ['jan','feb','march','april','may','june','july','august','september','october','november','december']
y = [1000,1200,1400,800,600,2400,400,500,2300,3400,3200,1000]
plt.xlabel('months')
plt.ylabel('sales amount')
plt.scatter(x,y,marker = '*')
plt.title('monthly sales report 2025')
plt.ylim(100,4000)
plt.xlim(0,11)
plt.grid(linestyle = ':',color = 'grey')
plt.legend()
plt.xticks([2,5,8,11],['q1','q2','q3','q4'])
plt.show()
```

```python
import matplotlib.pyplot as plt
product = ['a','b','c','d']
sales =[1500,1000,800,1200]
plt.barh(product,sales,color =  'green',label = 'product sales 2025')
plt.title('product wise sales in 2025')
plt.xlabel('product')
plt.ylabel('sales')
plt.legend()
plt.grid(linestyle=':',color='red')
plt.show()
```

```python
import matplotlib.pyplot as plt
regions = ['north','south','east','west']
revenue = [3000,1000,1500,2000]
plt.pie(revenue,labels = regions,autopct = '%1.2f%%',colors = ['gold','orange','crimson','red'])
plt.title('revenue contribution by regions')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
scores = [10,25,33,45,78,96,32,15,36,25,89,99,97,67,58,79,50]
plt.hist(scores,bins=10,color='crimson',edgecolor='black',label='score distribution')
plt.title('student score distribution across the score range')
plt.xlabel('score ranges')
plt.ylabel('no of students')
plt.legend()
plt.grid(linestyle=':',color='black')
plt.show()
```
