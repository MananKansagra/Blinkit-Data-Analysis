```python
import pandas as pd

# Clean column names (optional but helpful)
df.columns = df.columns.str.strip().str.lower()

# Group by target_audience and channel
grouped_df = df.groupby(['target_audience', 'channel'], as_index=False).agg({
    'clicks': 'sum',
    'impressions': 'sum',
    'conversions': 'sum',
    'spend': 'sum'
})

# Calculate CRPC (Cost per Conversion)
grouped_df['crpc'] = grouped_df.apply(
    lambda row: row['spend'] / row['conversions'] if row['conversions'] != 0 else 0,
    axis=1
)

# Round CRPC to 2 decimal places
grouped_df['crpc'] = grouped_df['crpc'].round(2)


# Show result
print(grouped_df)
```
