```python
df['pincode'] = df['pincode'].astype(str)
invalid_pincode_rows = df[df['pincode'].str.len() != 6]
print(invalid_pincode_rows)
```

```python
# Step 1: Ensure all pincodes are strings
df['pincode'] = df['pincode'].astype(str)

# Step 2: Identify rows where pincode is not exactly 6 digits
invalid_pincode_mask = df['pincode'].str.len() != 6

# Step 3: Drop those invalid rows
df = df[~invalid_pincode_mask]

# Step 4: Confirm the result
print("✅ Total rows after dropping invalid pincodes:", len(df))
print("All pincodes valid:", df['pincode'].str.len().eq(6).all())
```

```python
df.to_csv("pincode.csv",index = False)
```
