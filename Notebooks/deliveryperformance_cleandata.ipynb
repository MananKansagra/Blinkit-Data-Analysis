```python
import pandas as pd
df  = pd.read_csv('data/blinkit_delivery_performance.csv')
```

```python
df.isnull().sum()
```

```python
# Step 1: Count blank rows (NaN or empty string)
blank_rows = df['reasons_if_delayed'].isna().sum() + (df['reasons_if_delayed'] == '').sum()
print("Number of blank rows in 'reasons_if_delayed':", blank_rows)

# Step 2: Replace NaN and empty string with 'not applicable'
df['reasons_if_delayed'].replace(["", None], "Not Applicable", inplace=True)
df['reasons_if_delayed'].fillna("Not Applicable", inplace=True)
```
